```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

pd.pandas.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
# categorial features with nan value
features_with_na = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtype == 'O']
for feature in features_with_na:
    print('{} : {}% missing values'.format(feature, np.round(df[feature].isnull().mean(), 4)))
```

```python
# Replacing missing values with a new label

def replace_cat_feature(dataset, feature_with_nan):
    data = dataset.copy()
    
    data[feature_with_nan] = data[feature_with_nan].fillna('Missing')
    return data

df = replace_cat_feature(df, features_with_na)
df[features_with_na].isnull().sum()
```

```python
df.head()
```

```python
# Numerical variable that contains missing values
numericals_with_na = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtype != 'O']

for feature in numericals_with_na:
    print('{}: {}%missing value'.format(feature, np.round(df[feature].isnull().mean(), 4)))
```

```python
# Replacing the numerical missing values

for feature in numericals_with_na:
    median_value = df[feature].median()
    
    # create a new feature to capture nan value
    df[feature+'nan'] = np.where(df[feature].isnull(), 1, 0)
    df[feature].fillna(median_value, inplace = True)
    
df[numericals_with_na].isnull().sum()
    
```

```python
## Temporal variables (Data time variables)

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    
    df[feature] = df['YrSold'] - df[feature]
```

```python
import numpy as np
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    df[feature] = np.log(df[feature])
```

```python
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
categorical_features
```

```python
for feature in categorical_features:
    temp = df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df = temp[temp>0.01].index
    df[feature] = np.where(df[feature].isin(temp_df), df[feature], 'Rare_var')
    
```
