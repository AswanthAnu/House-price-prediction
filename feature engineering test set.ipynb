```python
import pandas as pd
import numpy as np
import import_ipynb
import House_prediction_feature_engineering as hp
```

```python
df = pd.read_csv('test.csv')
test_target = pd.read_csv('sample_submission.csv')
```

```python
df['SalePrice'] = test_target['SalePrice']
```

```python
df.head()
```

```python
# Replacing missing values with a new label
df = hp.replace_cat_feature(df, hp.features_with_na)
```

```python
df[hp.features_with_na].isnull().sum()
```

```python
# Replacing the numerical missing values

for feature in hp.numericals_with_na:
    median_value = df[feature].median()
    
    # create a new feature to capture nan value
    df[feature+'nan'] = np.where(df[feature].isnull(), 1, 0)
    df[feature].fillna(median_value, inplace = True)
    
df[hp.numericals_with_na].isnull().sum()
```

```python
## Temporal variables (Data time variables)

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    
    df[feature] = df['YrSold'] - df[feature]
```

```python
df.head()
```

```python
 df[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
df.head()
```

```python
for feature in hp.num_features:
    df[feature] = np.log(df[feature])
```

```python
df.head()
```

```python
for feature in hp.categorical_features:
    temp = df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df = temp[temp>0.01].index
    df[feature] = np.where(df[feature].isin(temp_df), df[feature], 'Rare_var')
    
```

```python
df.head()
```

```python
for feature in hp.categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
# Scaling feature
hp.scaler.transform(df[hp.feature_scale])
```

```python
data  = pd.concat([df[['Id', 'SalePrice']].reset_index(drop = True),
                  pd.DataFrame(hp.scaler.transform(df[hp.feature_scale]), columns=hp.feature_scale)], axis=1)
```
